```python
import numpy as np
import pandas as pd
dataset = pd.read_csv('data.csv')
print("Output: ",dataset.shape)
```

```python
#-------------------------------------------------------------------------------
dataset.info
```

```python
#-------------------------------------------------------------------------------
print("Output: ")
dataset.describe()
```

```python
#Dropping User Id and Gender Column
dataset = dataset.drop(['User ID','Gender'],axis=1)
print("Output: ")
dataset.head()
```

```python
#-------------------------------------------------------------------------------
#Checking Sum of Number of null values in each column
print("Output: ")
dataset.isnull().sum()
```

```python
#-------------------------------------------------------------------------------
#Adding mode values at null places of Country column
country_mode = dataset['country'].mode()[0]
dataset['country'].fillna(country_mode,inplace=True)
print("Output: ")
dataset.head()
```

```python
#-------------------------------------------------------------------------------
#Checking Sum of Number of null values in each column
print("Output: ")
dataset.isnull().sum()
```

```python
#-------------------------------------------------------------------------------
#Adding Numeric values instead of country names
df_cat = pd.get_dummies(dataset['country'],drop_first=False)
print("Output: ")
df_cat.head()
```

```python
#-------------------------------------------------------------------------------
#Concating numeric country columns with dataset and dropping country column
dataset = pd.concat([dataset,df_cat],axis=1)
dataset.drop('country',axis=1,inplace=True)
print("Output: ")
dataset.head()
```

```python
#-------------------------------------------------------------------------------
#Adding median values at the null places
from sklearn.impute import SimpleImputer
median_imputer = SimpleImputer(missing_values = np.nan,strategy = 'median')
result_median_imputer = median_imputer.fit_transform(dataset)
dataset = pd.DataFrame(result_median_imputer,columns = dataset.columns)
print("Output: ")
dataset.head()
```

```python
#-------------------------------------------------------------------------------
#Checking Sum of Number of null values in each column
print("Output: ")
dataset.isnull().sum()
```

```python
#-------------------------------------------------------------------------------
#Converting Datatype of France,India,USA,Purchased to int64
dataset['France'] = dataset['France'].astype('int64')
dataset['India'] = dataset['India'].astype('int64')
dataset['USA'] = dataset['USA'].astype('int64')
dataset['Purchased'] = dataset['Purchased'].astype('int64')
print("Output: ")
dataset.head()
```

```python
#-------------------------------------------------------------------------------
#Adding Target Column ie.Purchased Column to Extreme Right
purchase_column = dataset['Purchased']
dataset = dataset.drop('Purchased',axis=1)
dataset.insert(5,"Purchased",purchase_column)
print("Output: ")
dataset
```

```python
#-------------------------------------------------------------------------------
X=dataset.loc[:,['Age','salary','France','India','USA']]
y=dataset.loc[:,['Purchased']]
y=np.array(y)
y=y.ravel()
print("Output: ")
y
```

```python
#-------------------------------------------------------------------------------
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=41)
print("Output: ")
print("Total records(rows) in X_train:",len(X_train))
print("Total records(rows) in y_train:",len(y_train))
print("Total records(rows) in X_test:",len(X_test))
print("Total records(rows) in y_test:",len(y_test))
```

```python
#-------------------------------------------------------------------------------
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train=np.array(X_train)
X_test=np.array(X_test)
print("Output: ")
X_train[:5]
```
