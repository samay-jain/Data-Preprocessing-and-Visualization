```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
df = pd.read_csv('scores.csv')
```

```python
phy = df['Physics']
card = df['CardNo']
chem = df['Chemistry']
maths = df['Mathematics']
plt.plot(card,maths, color='red', linestyle='dashed', linewidth = 2,
         marker='o', markerfacecolor='blue', markersize=8)
plt.title("Line plot of CardNo and Score in Mathematics")
plt.legend(['Marks'],loc="upper left")
plt.xlabel("CardNo")
plt.ylabel("Score in Mathematics")
plt.grid(color="green",linewidth=1)
ax = plt.gca()
ax.set_aspect(0.3)
```

```python
df=df.head(10)
figure,ax = plt.subplots(figsize=(15,7))
maths = df['Mathematics']
chem = df['Chemistry']
phy = df['Physics']
name = df['Name']
plt.plot(name,maths,color='red',marker='o',markersize=8,label="Mathematics")
plt.plot(name,chem,color='green',marker='o',markersize=8,label="Chemistry")
plt.plot(name,phy,color='blue',marker='o',markersize=8,label="Physics")
plt.rc('font',size=10)
plt.legend(prop={'size':10},loc="upper left")
plt.grid(color="green")
plt.xlabel("Name")
plt.ylabel("Marks Obtained")
plt.title("Student with total Marks")
for index in range(len(name)):
  ax.text(name[index],maths[index]+2,maths[index],size=10,color="red")
  ax.text(name[index],chem[index]+2,maths[index],size=10,color="green")
  ax.text(name[index],phy[index]+2,maths[index],size=10,color="blue")
plt.show()
```
