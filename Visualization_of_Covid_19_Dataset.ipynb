```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('covid-19cases.csv')
```

```python
fig,ax = plt.subplots(figsize=(20,10))
ax.set_facecolor('0.8')
confCases = df['Daily Confirmed']
Date = df['Date']
plt.plot(Date,confCases,color="blue",marker='o',markersize=7,linewidth=4)
plt.title('Covid-19 Daily Confirmed Cases',size=20)
plt.xlabel('Date',size=20)
plt.ylabel('Confirmed covid-19 Cases',size=20)
plt.grid()
plt.legend(['Confirmed Cases'],loc='upper left')
for index in range(len(Date)):
  ax.text(Date[index],confCases[index]+8,confCases[index],size=14,color='black')
plt.ylim(0,4400)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(20,10))
ax.set_facecolor('0.8')
recovCases = df['Daily Recovered']
Date = df['Date']
plt.plot(Date,recovCases,color="blue",marker='o',markersize=7,linewidth=4)
plt.title('Covid-19 Daily Recovered Cases',size=20)
plt.xlabel('Date',size=20)
plt.ylabel('Recovered covid-19 Cases',size=20)
plt.grid()
plt.legend(['Recovered Cases'],loc='upper left')
plt.ylim(0,2000)
for index in range(len(Date)):
  ax.text(Date[index],recovCases[index]+8,recovCases[index],size=13,color='black')
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(20,10))
ax.set_facecolor('0.8')
dd = df['Daily Deceased']
Date = df['Date']
plt.plot(Date,dd,color="blue",marker='o',markersize=7,linewidth=4)
plt.title('Covid-19 Daily Deceased Cases',size=30)
plt.xlabel('Date',size=30)
plt.ylabel('Daily Deceased Cases',size=30)
plt.grid()
plt.legend(['Daily Deceased Cases'],loc='upper left')
for index in range(len(Date)):
  ax.text(Date[index],dd[index]+8,dd[index],size=14,color='black')
plt.ylim(0,145)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(20,10))
ax.set_facecolor('0.8')
plt.grid()
plt.plot(Date,confCases,color="red",marker='o',markersize=7,linewidth=4)
plt.plot(Date,recovCases,color="green",marker='o',markersize=7,linewidth=4)
plt.plot(Date,dd,color="blue",marker='o',markersize=7,linewidth=4)
plt.rc('font',size=10)
plt.legend(['Confirmed Cases','Recovered Cases','Daily Deceased'],loc="upper left")
plt.xlabel("Dates",size=20)
plt.ylabel("Number of Cases",size=20)
plt.title("Covid-19 Data Analysis",size=20)
for index in range(len(Date)):
  ax.text(Date[index],confCases[index]+8,confCases[index],size=12,color='black')
  ax.text(Date[index],recovCases[index]+8,recovCases[index],size=12,color='black')
  ax.text(Date[index],dd[index]+8,dd[index],size=12,color='black')
plt.show()
```
