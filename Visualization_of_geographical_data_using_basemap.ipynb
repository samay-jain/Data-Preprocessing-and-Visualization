```python
plt.figure(figsize=(8,8))
m=Basemap(projection='ortho',resolution='f',lat_0=25,lon_0=-285)
m.bluemarble(scale=0.5);
```

```python
fig=plt.figure(figsize=(8,8))
m=Basemap(projection='lcc',resolution='f',width=8E6,height=8E6,lat_0=20.5937,lon_0=78.9629)
m.etopo(scale=0.5,alpha=0.5)
x,y=m(75,22)
#plt.plot(x,y,'ok',markersize=5)
plt.text(x,y,'India',fontsize=15)
```

```python
from itertools import chain
all_lines=set()
def draw_map(m,scale=0.5):
  m.shadedrelief(scale=scale)
  lats=m.drawparallels(np.linspace(-90,90,13))
  lons=m.drawmeridians(np.linspace(-180,180,13))
  lat_lines=chain(*(tup[1][0] for tup in lats.items()))
  lon_lines=chain(*(tup[1][0]for tup in lons.items()))
  all_lines=chain(lat_lines,lon_lines)
for line in all_lines:
  line.set(linestyle='-',aplha=0.9,color='w')
fig=plt.figure(figsize=(8,6),edgecolor='w')
m=Basemap(projection='cyl',resolution='f',llcrnrlat=-90,urcrnrlat=90,llcrnrlon=-180,urcrnrlon=180)
draw_map(m)
```

```python
fig = plt.figure(figsize=(8, 6), edgecolor='b')
m = Basemap(projection='moll', resolution='f',lat_0=0, lon_0=0)
draw_map(m)
```

```python
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='ortho', resolution='f',lat_0=50, lon_0=50)
draw_map(m);
```

```python
fig = plt.figure(figsize=(8, 8))
m = Basemap(projection='lcc', resolution='f',lon_0=80, lat_0=38, lat_1=70,
            lat_2=10,width=1.6E7, height=1.2E7)
draw_map(m)
```
