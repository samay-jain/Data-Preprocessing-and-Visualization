```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
feature_x = np.arange(0, 50, 2)
feature_y = np.arange(0, 50, 3)
[X, Y] = np.meshgrid(feature_x, feature_y)
Z = np.cos(X/2) + np.sin(Y/4)
plt.contour(X, Y, Z,colors='black')
plt.colorbar()
plt.show()
```

```python
plt.contourf(X,Y,Z,20,cmap='RdGy')
plt.colorbar()
plt.show()
```

```python
plt.imshow(Z,extent=[0,5,0,5],origin='lower',cmap='RdGy')
plt.colorbar()
plt.show()
```

```python
contours = plt.contour(X,Y,Z,6,colors='black')
plt.clabel(contours,inline=True,fontsize=8)
plt.imshow(Z,extent=[0,48,0,48],origin='lower',cmap='RdGy',alpha=0.5)
plt.colorbar()
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#sns.get_dataset_names()
df = sns.load_dataset('anagrams')
df = df.dropna()
df
```

```python
feature_x=df['subidr']
feature_y=df['num2']
[X, Y] = np.meshgrid(feature_x,feature_y)
Z = np.sin(X/2) + np.cos(Y)
plt.rcParams['figure.figsize'] = [10,5]
plt.contourf(X, Y, Z,15,cmap='rainbow')
plt.xlabel("x-axis",fontsize=15)
plt.ylabel("y-axis",fontsize=15)
plt.title("Contour plot of Anagrams Dataset",fontsize=20)
plt.colorbar()
plt.show()
```
